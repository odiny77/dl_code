```python
import numpy as np
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 3)# For reproducibility
```

```python
true_W = np.array([[3], [2], [-1.5]])
```

```python
y = X.dot(true_W) + np.random.randn(100, 1)
```

```python
W = np.random.randn(X.shape[1], 1)
```

```python
learning_rate = 0.001
n_iterations = 10000
m = len(y)
```

```python
m
```

```python
for iteration in range(n_iterations):
    gradients = -1/m * X.T@(y - X@W)
    W = W - learning_rate * gradients
    if iteration % 100 == 0:
        loss = np.mean((y - X@W) ** 2 ) / 2
        print(f"Iteration {iteration}: Loss = {loss:.4f}")
```

```python
print("True parameters:", true_W)
print("Optimal parameters:", W)
```
